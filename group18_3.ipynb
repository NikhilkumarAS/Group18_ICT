```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Vimal\Desktop\DAS\week 3\diabetes_model.csv")
```

```python
df.head()
```

# Create a Count plot of Outcome

```python
plt.figure(figsize=(10,8))
ax=sns.countplot(x=df.Outcome)
plt.title("Count plot of Outcome")

for p in ax.patches:
    ax.annotate("{:0f}".format(p.get_height()),(p.get_x()+0.25,p.get_height()+2))
    
plt.show()
```

# Analyze the effect of age on Outcome using Boxplot

```python
plt.figure(figsize=(10,8))
sns.boxplot(x="Outcome",y="Age",data=df)
xlabel='Outcome'
ylabel='Age'
plt.title("Relationship Of Age with Outcome")
plt.show()
```

# getting diabetes if it is common in family, using the Daiabetes Pedigree Function

```python
plt.figure(figsize=(10,8))
#sns.scatterplot(y="DiabetesPedigreeFunction",x="Outcome",data=df)
sns.boxplot(x="Outcome",y="DiabetesPedigreeFunction",data=df)
xlabel='Outcome'
ylabel='DiabetesPedigreeFunction'
plt.title("Relationship Of DPF with Outcome")
plt.show()
```

# Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
plt.figure(figsize=(10,10))
#sns.boxplot(x="Glucose",y="Insulin",data=df)
sns.scatterplot(x="Glucose",y="Insulin",data=df,hue="Outcome")
xlabel='Glucose'
ylabel='Insulin'
plt.title("Relationship Of Insulin & Glucose")
plt.show()
```

# Create a heatmap on Correlation of features in the data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),cmap="Blues",annot=True)
plt.title("Heatmap")
plt.show()
```
